```python
data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
data.head()
```

```python
column_to_drop = 'User_ID'
data = data.loc[:, data.columns != column_to_drop]
```

```python
print(data)
```

```python
x=data.drop(columns=['Calories'], axis=1)
y=data['Calories']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2) #splitting the into train and split data
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score #it is used find the accuracy of our model
```

```python

# loading the Linear Regression model
model=LinearRegression()
model.fit(x_train,y_train) #fitting the model
y_pred=model.predict(x_test)
r2= metrics.r2_score(y_test,y_pred)
print('accuracy of the model is: ', r2)
```

```python
data_size=data.size
print(data_size)
```

```python
num_columns = data.shape[1]
num_columns
```

```python
adj_r2 = 1 - ((1 - r2) * (data_size - 1)) / (data_size -num_columns  - 1)
print("the adjusted r2 value is",adj_r2)
```

evaluating the model

```python
# 1st INPUT
inputdata= (1,34,179,71,13,100,40.5) #female,34,179,71,13,100,40.5,71
input_array=np.asarray(inputdata)# changing input_data to a numpy array
input=input_array.reshape(1,-1) # reshape the array
prediction=model.predict(input)
print("she burnt",prediction,"calories")
```

```python
prediction2=model.predict(np.asarray((0,31,182,83,20,101,40.9)).reshape(1,-1)) #male,31,182,83,20,101,40.9,107
print("he burnt",prediction2,"calories of energy")
```
