```python
#separate the data and label them as X and Y
x=df.iloc[:,0:8] #retrive all the rows and columns from 0 to 8 ,8th column not included
```

```python
print(x)
```

```python
y=df["Outcome"] #gives only 8th column i.e outcomes
```

```python
print(y)
```

```python
#importing the dependencies
import numpy as np
from sklearn.preprocessing import StandardScaler #this standard  scalar is used to standardize the data to a common range
from sklearn.model_selection import train_test_split #to split our data into trainng and testing data
from sklearn import svm #SVM stands for support vector machine
from sklearn.metrics import accuracy_score #it is used find the accuracy of our model
```

```python
#Data Standardization-standardizing data to a common range will help to our machine learning model to make better predictions
```

```python
scaler = StandardScaler() #loading standard scalar function into scaler
```

```python
scaler=scaler.fit(x)  #fitting the data into scaler
```

```python
standardized_data=scaler.transform(x)
```

```python
print(standardized_data) #all the values of the data will be in a commom range i.e 0 to 1
```

```python
x=standardized_data
```

```python
#train test split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, stratify=y, random_state=2)
# test size represents how much of data you wants to give for testing the model,here 0.2 represents 20% of testing data
#stratify method is used to for separting same proportion y data for testing and traing sets
#random state is used for controlling the shuffling process
```

Radial Basis Function (RBF) Kernel:

```python
# Create an SVC classifier
model =svm.SVC(kernel='rbf')

# Train the model
model.fit(x_train, y_train)

# Make predictions on the test set
y1_pred = model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y1_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y1_pred)
print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, y1_pred))
```

Polynomial Kernal

```python
model2 =svm.SVC(kernel='poly')
model2.fit(x_train, y_train)
y2_pred = model2.predict(x_test)
accuracy = accuracy_score(y_test, y2_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y2_pred)
print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, y2_pred))
```

sigmoid Kernal

```python
model3 =svm.SVC(kernel='sigmoid')
model3.fit(x_train, y_train)
y3_pred = model3.predict(x_test)
accuracy = accuracy_score(y_test, y3_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y3_pred)
print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, y3_pred))
```

Linear Kernal

```python
model4 =svm.SVC(kernel='linear')
model4.fit(x_train, y_train)
y4_pred = model4.predict(x_test)
accuracy = accuracy_score(y_test, y4_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y4_pred)
print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, y4_pred))
```
